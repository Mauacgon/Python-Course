```python
import random

suits = ('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two', "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace")
values = {'Two':2, "Three":3, "Four":4, "Five":5, "Six":6, "Seven":7, "Eight":8,
        "Nine":9, "Ten":10, "Jack":10, "Queen":10, "King":10, "Ace":1}

class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
    
class Deck():
    
    def __init__(self):
        self.all_cards = []
        
        for i in suits:
            for j in ranks:
                self.all_cards.append(Card(i,j))
                
    def add_cards(self, new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
                
    def shuffle(self):
        
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        
        self.all_cards.pop()
        
class Player():
    
    def __init__(self,name,coins):
        self.name = name
        self.coins = coins
        self.all_cards = []
        
    def add_coins(self, number_of_coins):
        self.coins = self.coins + number_of_coins
        
    def deal_one(self):
        self.all_cards.pop()
        
    def deal_all(self):
        self.all_cards = []
        
    def add_cards(self, new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
            
    def total_value(self):
        values = [i.value for i in self.all_cards]
        ranks = [i.rank for i in self.all_cards]
        
        v_sum = sum(values)
        
        if v_sum < 21 and v_sum + 10 < 21:
            if 'Ace' in ranks:
                v_sum = v_sum + 10
                
        return v_sum          
            
            
class Bankroll():
    
    def __init__(self):
        self.all_cards = []
     
    def deal_one(self):
        self.all_cards.pop()
        
    def deal_all(self):
        self.all_cards = []
        
    def add_cards(self, new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
            
    def total_value(self):
        values = [i.value for i in self.all_cards]
        ranks = [i.rank for i in self.all_cards]
        
        v_sum = sum(values)
        
        if v_sum < 21 and v_sum + 10 < 21:
            if 'Ace' in ranks:
                v_sum = v_sum + 10
                
        return v_sum
    
        
```

```python
name = input("Choose a name for player one______")

true_deposit = True
    
while true_deposit:
    try:
        coins = int(input(f'Player 1 name is {name}, now choose a deposit of coins for {name} different than 0 _____'))
        true_deposit = False
        break
    except ValueError:
        print("Enter a valid integer")
    

if coins == 0:
    while coins <= 0:
        coins = int(input(f'Deposit must be higher than 0, please, retype the choosen quantity of the deposit ___'))

Player1 = Player(name, coins)

Deck1 = Deck()

Bankroll1 = Bankroll()                  

carry_on = True

while carry_on:
    
    if Player1.coins == 0:
        carry_on = False
        break
    
    Deck1.shuffle()
    Deck1.shuffle()

    print("Deck has been shuffled")
    
    right_bet = True
    
    while right_bet:
        try:
            bet = int(input('place an inital bet____'))
            right_bet = False
            break
        except:
            print("Bet must be an integer, please enter a valide bet")
    
    while bet > Player1.coins or bet <= 0:
        try:
            bet = int(input('bet cannot be higher than the number of coins in the bankroll, please, retype the choosen bet ___'))
        except ValueError:
            print("Bet must be an integer, please enter a valide bet")
                  
    Player1.add_coins(-bet)

    input('press some key to deal the cards____')

    for i in range(1,3):
        Player1.add_cards(Deck1.all_cards[-sum(range(0,i+1))])
        Bankroll1.add_cards(Deck1.all_cards[-sum(range(0,i+1))-1])
        
    for i in range(0,4):
        Deck1.deal_one()
    
    carry_on_bet = True
    
    while carry_on_bet == True:
        
        if Player1.total_value() > 21:
            print(f'Player1 bursts, lost bet, total {Player1.name} coins are {Player1.coins}')
            Deck1.add_cards(Player1.all_cards)
            Deck1.add_cards(Bankroll1.all_cards)
            Player1.deal_all()
            carry_on_bet = False            
            break
        
        print(f'{Player1.name} cards are:')
        print(Player1.all_cards[0])
        print(Player1.all_cards[1])
        print(f'Bankroll first card is:')
        print(Bankroll1.all_cards[0])

        possible_actions = ("Hit", "Stay")

        action = ''
        
        Bankroll_hit = True

        while action != 'Hit' and action != 'Stay':
            action = input(f'Your cards have a total value of {Player1.total_value()}\
             choose an action between Hit or Stay ___')

        if action == 'Hit':
            Player1.add_cards(Deck1.all_cards[-1])
            Deck1.deal_one()
            
        if action == 'Stay':
            while Bankroll_hit == True:
                
                if Bankroll1.total_value() > Player1.total_value() and Bankroll1.total_value() <= 21:
                    print(f'Bankroll total value is {Bankroll1.total_value()}')
                    print("Bankroll wins, lost bet")
                    Deck1.add_cards(Player1.all_cards)
                    Deck1.add_cards(Bankroll1.all_cards)
                    Player1.deal_all()
                    Bankroll1.deal_all()
                    Bankroll_hit = False
                    carry_on_bet = False
                    input("press a Key to continue ____")
                    break
                    
                if Bankroll1.total_value() > 21:
                    Player1.add_coins(bet*2)
                    print(f'Player1 wins, won bet, total {Player1.name} coins are {Player1.coins}')
                    Deck1.add_cards(Player1.all_cards)
                    Deck1.add_cards(Bankroll1.all_cards)  
                    Player1.deal_all()
                    Bankroll1.deal_all()
                    Bankroll_hit = False
                    carry_on_bet = False
                    input("press a Key to continue ____")
                    break
                    
                if Bankroll1.total_value() <= Player1.total_value():
                    Bankroll1.add_cards(Deck1.all_cards[-1])
                    Deck1.deal_one()
                    print(f'Bankroll ads a card, its total value is {Bankroll1.total_value()}')
                    input("press a Key to continue ____")
                    
                if Bankroll1.total_value() == Player1.total_value() and Bankroll1.total_value() == 21:
                    print("Draw, nobody wins")
                    Deck1.add_cards(Player1.all_cards)
                    Deck1.add_cards(Bankroll1.all_cards)                    
                    Bankroll_hit = False
                    carry_on_bet = False
                    input("press a Key to continue ____")
                    break
                    
print(f'{Player1.name} has run out of coins, GAME OVER')
    
```
